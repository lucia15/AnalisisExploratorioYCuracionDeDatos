```python
import matplotlib as plt
import pandas as pd
import numpy as np
import seaborn as sb
```

```python
tabla_1 = pd.read_csv("https://github.com/DiploDatos/AnalisisYCuracion/raw/master/input/wtageinf.csv")
tabla_1.head(10)
```

```python
filtrado= tabla_1[tabla_1.Sex==1]
filtrado.plot(x='Agemos', y='P3');
```

```python
peso = tabla_1
```

```python
#Peso para Niños
peso_ninios = peso[peso.Sex==1]
peso_ninios.plot(x='Agemos', y=['P3','P5','P10','P25','P50','P75','P90','P95','P97'])
plt.pyplot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.pyplot.title("Peso de Niños")
sb.despine()
```

```python
#Peso para Niñas
peso_ninias = peso[peso.Sex==2]
peso_ninias.plot(x='Agemos', y=['P3','P5','P10','P25','P50','P75','P90','P95','P97'])
plt.pyplot.legend(bbox_to_anchor=(1.05, 1), loc=2, borderaxespad=0.)
plt.pyplot.title("Peso de Niñas")
sb.despine()
```
