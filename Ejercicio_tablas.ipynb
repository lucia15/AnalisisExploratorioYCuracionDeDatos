```python
tabla_2 = pd.read_json("https://github.com/DiploDatos/AnalisisYCuracion/raw/master/input/pasos.json")
tabla_2.head(3)
```

```python
Provincias = tabla_2.province.unique() 
for p in sorted(Provincias):
    print(p)
```

Es posible observar que Neuquén y Córdoba se encuentran 2 veces en la lista. Además, figura como Provincia el nombre del AEROPUERTO INTERNACIONAL PILOTO CIVIL NORBERTO FERNANDEZ.  A continuación, procedemos a corregir esto.

```python
tabla_2[tabla_2.province == 'NEUQUEN'] = 'Neuquen'
tabla_2[tabla_2.province == 'Cordoba'] = 'Córdoba'
tabla_2[tabla_2.province == 'AEROPUERTO INTERNACIONAL PILOTO CIVIL NORBERTO FERNANDEZ'] = "Santa Cruz"
Provincias = tabla_2.province.unique() 
for p in sorted(Provincias):
    print(p)
```

```python
pypt.figure(figsize=(20,6))
ax = sb.countplot(data=tabla_2, x='province', order = tabla_2.province.value_counts().index, palette="Set3")
ax.grid(axis='y')
pypt.title("Pasos Fronterizos en cada Provincia de Argentina")
pypt.ylabel("Número de Pasos Fronterizos")
pypt.xlabel("Provincias de Argentina")
sb.despine()
pypt.xticks(rotation=45)
```
